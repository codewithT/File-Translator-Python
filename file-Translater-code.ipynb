```python
 
#from tkinter import *
import tkinter as tk
import googletrans
import textblob
from tkinter import ttk,messagebox
rot= tk.Tk()
rot.title('File language Translator')
rot.geometry("1300x1275")
rot.iconbitmap("C:\\Users\\hi\\Downloads\\language_learn_icon-icons.com_51070.ico")
rot.grid_columnconfigure((0,1,2,3), weight=2)
rot.grid_rowconfigure((0,1,2,3), weight=1)
rot.configure(bg='#33FFFF')
#functions:
def translater():
    try:
        #To delete the data translated in past
        translated_text.delete(1.0,tk.END)
        #get languages from dictionary keys
        #get the from language key
         
        for key,value in language.items():
            if(value==original_combo.get()):
                from_language_key = key
         #get to language key
        for key,value in language.items():
            if(value==translated_combo.get()):
                to_language_key = key
        
        file = open(input_box.get(),"r")
        contents = file.read()
        #turn original text into a textblob
        
        words =textblob.TextBlob(contents)
        #translate text
        words = words.translate(from_lang = from_language_key,to =to_language_key)
        #output translated text to screen
        translated_text.insert(1.0,words)
        
        
    except Exception as e:
        messagebox.showerror("translator",e)
## creating another functions::
def translate_it():
    
    
    translated_text1.delete(1.0,tk.END)
    try:
        
        #get languages from dictionary keys
        #get the from language key
        for key,value in language.items():
            if(value==original_combo.get()):
                from_language_key = key
         #get to language key
        for key,value in language.items():
            if(value==translated_combo.get()):
                to_language_key = key
        #turn original text into a textblob
        words1 =textblob.TextBlob(original_text.get(1.0,tk.END))
        #translate text
        words1= words1.translate(from_lang = from_language_key,to =to_language_key)
        #output translated text to screen
        translated_text1.insert(1.0,words1)
        
        
    except Exception as e:
        messagebox.showerror("translator",e)
#creating a button to translate language.
T_button =tk.Button(rot,text = "Translate File",font= ("Helvetica",24),command =translater)
T_button.grid(row =0,column =1)
#creating a button to translate language.
T_button1 =tk.Button(rot,text = "Translate!",font= ("Helvetica",24),command =translate_it)
T_button1.grid(row =2,column =1)


#grab language list from google trans
language= googletrans.LANGUAGES
#convert to list
language_list = list(language.values())
#translated boxes::***************

translated_text = tk.Text(rot,height = 15,width =50)
#translated_text.grid(row= 0,column = 2,pady =10,padx=10)
translated_text.place(x =800,y=75)
translated_text.configure(bg="#E0E0E0")
#label for translated_text box::***********
textbox1 =tk.Label(rot,text ="Translated Text:",font= ("Helvetica",24))
textbox1.place(x =800,y=10)
#translated boxes::***************
translated_text1 = tk.Text(rot,height = 10,width =40)
translated_text1.grid(row= 2,column = 2,pady =10,padx=10)
translated_text1.configure(bg="#E0E0E0")
 
#text boxes
original_text = tk.Text(rot,height = 10,width =40)
original_text.grid(row= 2,column = 0,pady =20,padx=20)
original_text.configure(padx=10)
original_text.configure(bg="#E0E0E0")
#original_text.place(x=50,y=400)
                   
##input field
label_file =tk.Label(rot,text ="Enter the file name:",font= ("Helvetica",24))
#label_file.grid(row = 0,column=0,padx =0,pady =0)
label_file.place(x =100,y=30)
input_box =tk.Entry(rot)
input_box.grid(row=0,column =0)
 

#combo boxes
original_combo = ttk.Combobox(rot,width = 50,value = language_list)
original_combo.current(21)
#original_combo.grid(row =1,column =0,padx= 20,pady = 20)
original_combo.place(x =50,y =200)
#combobox label:
combo_label =tk.Label(rot,text ="Select From language:",font= ("Helvetica",24))
combo_label.place(x =50,y =150)

translated_combo = ttk.Combobox(rot,width = 50,value = language_list)
translated_combo.current(26)
translated_combo.place(x =50,y =300)
#translated_combo boxes
combo_label2 =tk.Label(rot,text ="Select To language:",font= ("Helvetica",24))
combo_label2.place(x =50,y =250)

###
#clear funtions
def clear():
    translated_text.delete(1.0,tk.END)
    translated_text1.delete(1.0,tk.END)

#clear buttons
clear_button = tk.Button(rot,text ="Clear",command = clear)
clear_button.grid(row =1,column =1,padx = 10,pady=10)
clear_button.configure(bg='#CCE5FF')


rot.mainloop()
 
```
